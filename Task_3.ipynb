1) Построить кривую Безье второго порядка. Далее, подвергнуть композиции следующих афинных преобразований: перенос, сдвиг в квадрате, поворот на $\Pi / 4$

```python
# Задание 1

import numpy as np
import matplotlib.pyplot as plt

def bezier_linear(P0, P1, t):
    return (1 - t) * P0 + t * P1 # Берем формулки из Википедии https://ru.wikipedia.org/wiki/Кривая_Безье

A = np.array([-10, -7, 1])
B = np.array([15, 23, 1])

t = np.linspace(0, 1, 100)

curve = np.array([bezier_linear(A, B, ti) for ti in t])

plt.plot(curve[:, 0], curve[:, 1], label='Отрезок AB (кривая Безье)')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Отрезок AB, построенный кривой Безье')
plt.legend()
plt.show()
```

```python
# Задание 2.1

def bezier_quadratic(P0, P1, P2, t):
    return (1 - t)**2 * P0 + 2 * (1 - t) * t * P1 + t**2 * P2

P0 = np.array([0, 0, 1])
P1 = np.array([5, 20, 1])
P2 = np.array([10, 0, 1])

curve = np.array([bezier_quadratic(P0, P1, P2, ti) for ti in t])

def translate(point, dx, dy):
    translation_matrix = np.array([
        [1, 0, dx],
        [0, 1, dy],
        [0, 0, 1]
    ])
    return np.dot(translation_matrix, point)

def scale(point, sx):
    scaling_matrix = np.array([
        [1, sx, 0],
        [0, 1, 0],
        [0, 0, 1]
    ])
    return np.dot(scaling_matrix, point)

def rotate(point, angle):
    rotation_matrix = np.array([
        [np.cos(angle), -np.sin(angle), 0],
        [np.sin(angle), np.cos(angle), 0],
        [0, 0, 1]
    ])
    return np.dot(rotation_matrix, point)

translated_curve = np.array([translate(p, 5, 5) for p in curve])
scaled_curve = np.array([scale(p, 2) for p in translated_curve])
rotated_curve = np.array([rotate(p, np.pi / 4) for p in scaled_curve])

plt.plot(curve[:, 0], curve[:, 1], label='Исходная кривая')
plt.plot(rotated_curve[:, 0], rotated_curve[:, 1], label='Преобразованная кривая')
plt.plot([P0[0], P1[0], P2[0]], [P0[1], P1[1], P2[1]], 'ro--', label='Опорные точки')
plt.grid(True)
plt.legend()
plt.title('Кривая Безье второго порядка с аффинными преобразованиями')
plt.show()
```

```python
# Задание 2.2

def bezier_cubic(P0, P1, P2, P3, t):
    return (1 - t)**3 * P0 + 3 * (1 - t)**2 * t * P1 + 3 * (1 - t) * t**2 * P2 + t**3 * P3

P0 = np.array([0, 0, 1])
P1 = np.array([0, 5, 1])
P2 = np.array([5, 5, 1])
P3 = np.array([5, 0, 1])

curve = np.array([bezier_cubic(P0, P1, P2, P3, ti) for ti in t])

plt.plot(curve[:, 0], curve[:, 1], label='Кривая Безье третьего порядка')
plt.plot([P0[0], P1[0], P2[0], P3[0]], [P0[1], P1[1], P2[1], P3[1]], 'ro--', label='Опорные точки')
plt.grid(True)
plt.legend()
plt.title('Кривая Безье третьего порядка')
plt.show()
```
